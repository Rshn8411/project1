```python
# Creating Total Costs per state dfs (formating them as well)
   
# 2011 total cost per state df
medicare_total_cost_2011 = medicare_state_data_2011.iloc[:, [0, 4, 16]]
medicare_total_cost_2011['Total Actual Costs'] = medicare_total_cost_2011[
    'Total Actual Costs'].str.replace(',', '').str.replace('$', '').astype(float)
medicare_total_cost_2011['FFS Beneficiaries'] = (
                medicare_total_cost_2011['FFS Beneficiaries'
                ].str.replace(',','').astype(int))

# 2012 total cost per state df
medicare_total_cost_2012 = medicare_state_data_2012.iloc[:, [0, 4, 16]]
medicare_total_cost_2012['Total Actual Costs'] = medicare_total_cost_2012[
    'Total Actual Costs'].str.replace(',', '').str.replace('$', '').astype(float)
medicare_total_cost_2012['FFS Beneficiaries'] = (
                medicare_total_cost_2012['FFS Beneficiaries'
                ].str.replace(',','').astype(int))

# 2013 total cost per state df
medicare_total_cost_2013 = medicare_state_data_2013.iloc[:, [0, 4, 16]]
medicare_total_cost_2013['Total Actual Costs'] = medicare_total_cost_2013[
    'Total Actual Costs'].str.replace(',', '').str.replace('$', '').astype(float)
medicare_total_cost_2013['FFS Beneficiaries'] = (
                medicare_total_cost_2013['FFS Beneficiaries'
                ].str.replace(',','').astype(float))
    
# 2014 total cost per state df
medicare_total_cost_2014 = medicare_state_data_2014.iloc[:, [0, 4, 16]]
medicare_total_cost_2014['Total Actual Costs'] = medicare_total_cost_2014[
    'Total Actual Costs'].str.replace(',', '').str.replace('$', '').astype(float)
medicare_total_cost_2014['FFS Beneficiaries'] = (
                medicare_total_cost_2014['FFS Beneficiaries'
                ].str.replace(',','').astype(int))

# 2015 total cost per state df
medicare_total_cost_2015 = medicare_state_data_2015.iloc[:, [0, 4, 16]]
medicare_total_cost_2015['Total Actual Costs'] = medicare_total_cost_2015[
    'Total Actual Costs'].str.replace(',', '').str.replace('$', '').astype(float)
medicare_total_cost_2015['FFS Beneficiaries'] = (
                medicare_total_cost_2015['FFS Beneficiaries'
                ].str.replace(',','').astype(int))

# 2016 total cost per state df
medicare_total_cost_2016 = medicare_state_data_2016.iloc[:, [0, 4, 16]]
medicare_total_cost_2016['Total Actual Costs'] = medicare_total_cost_2016[
    'Total Actual Costs'].str.replace(',', '').str.replace('$', '').astype(float)
medicare_total_cost_2016['FFS Beneficiaries'] = (
                medicare_total_cost_2016['FFS Beneficiaries'
                ].str.replace(',','').astype(int))
```

```python
# Merge 2011 and 2012 total costs per state
medicare_total_cost_early = pd.merge(medicare_total_cost_2011, medicare_total_cost_2012, 
                        on="State", suffixes=(" 2011", " 2012"))

# Merge 2013 and 2014 total costs per state
medicare_total_cost_mid = pd.merge(medicare_total_cost_2013, medicare_total_cost_2014, 
                        on="State", suffixes=(" 2013", " 2014"))

# Merge 2015 and 2016 total costs per state
medicare_total_cost_late = pd.merge(medicare_total_cost_2015, medicare_total_cost_2016, 
                        on="State", suffixes=(" 2015", " 2016"))

# Merge all total costs per state
medicare_state_merge1 = pd.merge(medicare_total_cost_early, 
                            medicare_total_cost_mid, on="State")

# Merge all total costs per state
medicare_state_costs = pd.merge(medicare_state_merge1, 
                            medicare_total_cost_late, on="State")

# Sorted values
medicare_state_costs = medicare_state_costs.sort_values(
                            by = 'Total Actual Costs 2011', ascending = False)

# Reformatted df
pd.options.display.float_format = '{:,.0f}'.format
medicare_state_costs = medicare_state_costs.rename(
                            columns = {'FFS Beneficiaries 2011':'Beneficiaries 2011',
                            'FFS Beneficiaries 2012':'Beneficiaries 2012',
                            'FFS Beneficiaries 2013':'Beneficiaries 2013',           
                            'FFS Beneficiaries 2014':'Beneficiaries 2014',
                            'FFS Beneficiaries 2015':'Beneficiaries 2015',
                            'FFS Beneficiaries 2016':'Beneficiaries 2016'})

# Save to csv
medicare_state_costs.to_csv('Clean_2011_2016/medicare_state_costs2011_2016.csv', 
                                        index = True, header = True)

medicare_state_costs.style.format({'Total Actual Costs 2011':'${:,.2f}'.format,
                        'Total Actual Costs 2012':'${:,.2f}'.format,
                        'Total Actual Costs 2013':'${:,.2f}'.format,
                        'Total Actual Costs 2014':'${:,.2f}'.format,
                        'Total Actual Costs 2015':'${:,.2f}'.format,
                        'Total Actual Costs 2016':'${:,.2f}'.format, 
                        'Beneficiaries 2011': '{:,.0f}'.format,
                        'Beneficiaries 2012': '{:,.0f}'.format,
                        'Beneficiaries 2013': '{:,.0f}'.format,
                        'Beneficiaries 2014': '{:,.0f}'.format,
                        'Beneficiaries 2015': '{:,.0f}'.format,
                        'Beneficiaries 2016': '{:,.0f}'.format,})
```
