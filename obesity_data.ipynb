```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
healthcomp_csv = pd.read_csv('Clean_2011_2016/health_consumption_expenditures2011_2016.csv')
total_national = pd.read_csv('Clean_2011_2016/total_national_expenditures2011-216.csv')
historical = pd.read_csv('Clean_2011_2016/healthcare_historical_data_summary2011_2016.csv')
health_habits = pd.read_csv('Resources/health_habits.csv')

healthConsumptionDf = pd.DataFrame(healthcomp_csv)
totalExpenditures = pd.DataFrame(total_national)
historicalHealthCare = pd.DataFrame(historical)
health_behavior = pd.DataFrame(health_habits)
```

```python
health_behavior.rename(columns={"Start":"Year","LocationAbbr":"State","Data_Value":"Percent Adults"},inplace=True)
clean_health_behavior = health_behavior.loc[:,["Year","State","Topic","Question","Percent Adults","Sample_Size","Total","GeoLocation"]]
#clean_health_behavior.head()
```

```python
clean_health_behavior['Obese_Count'] = round((clean_health_behavior['Percent Adults']/100) * clean_health_behavior['Sample_Size'],2)
```

```python
clean_health_behavior.head()
```

```python
total_by_state = clean_health_behavior.loc[clean_health_behavior["Total"] == "Total"]
total_obesity_by_state = total_by_state.loc[total_by_state["Question"] == "Percent of adults aged 18 years and older who have obesity"]
us_total = total_obesity_by_state.loc[total_obesity_by_state["State"] == "US"]
total_obesity_by_state.set_index("State",inplace=True)
total_obesity_by_state.sort_values(by=['Year'],inplace=True)
total_obesity_by_state.head()
```

```python
drop_rows = total_obesity_by_state.drop(["VI","GU","PR"],axis=0)
```

```python
overall_cleaned_obesity_data = total_obesity_by_state.loc[:,["Year","Percent Adults","Sample_Size","Obese_Count","GeoLocation"]] 
```

```python
overall_cleaned_obesity_data.to_csv('overall_clean_obesity_data.csv')
```

```python
us_total = us_total.loc[:,["Year","Percent Adults","Sample_Size","Obese_Count"]] 
us_total.sort_values(by=['Year'],inplace=True)
us_total.head()
```
