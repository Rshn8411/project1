```python
import os
import requests
import json
from pprint import pprint
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# file paths
path_age_2000s = ('Cleaned_Data/age_distribution_2000s.csv')
#path_age_2011 = ('Cleaned_Data/age_distribution_2011.csv')
#path_age_2012 = ('Cleaned_Data/age_distribution_2012.csv')
#path_age_2013 = ('Cleaned_Data/age_distribution_2013.csv')
#path_age_2014 = ('Cleaned_Data/age_distribution_2014.csv')
#path_age_2015 = ('Cleaned_Data/age_distribution_2015.csv')
#path_age_2016 = ('Cleaned_Data/age_distribution_2016.csv')
```

```python
# display data, remove unnamed index.
age_dist_2000s = pd.read_csv(path_age_2000s)
age_dist_2000s = age_dist_2000s.rename(columns={'Unnamed: 0': 'Age'})
#age_dist_2000s = age_dist_2000s.set_index('Age')
age_dist_2000s
```

```python
year_2000 = pd.DataFrame(age_dist_2000s[['Age', '2000']])

year_2000
```

```python
year_2000 = year_2000.rename(columns={'2000': 'Population'})

year_2000
```

```python
age_groups = ['18 to 24', '25 to 34', '35 to 44', '45 to 54', '55 to 64', '65 and over']
#objects = year_2000['Age']
#population = [19116667, 39804497, 45168843, 37998974, 24429066, 35069568]
population = [19, 39, 45, 37, 24, 35]
#age = year_2000.Age

plt.bar(age_groups, population, color='r', align='center')

plt.ylim(0, max(population)+5)
plt.xlim(-0.75, len(age_groups)-0.25)
tick_locations = [value for value in age_groups]
plt.xticks(tick_locations, age_groups)
plt.xlabel("Age Groups")
plt.ylabel("Population in Millions")
plt.title("Age Distribution in 2000")
```
