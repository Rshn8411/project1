```python
# Dependencies
import os
import requests
import json
from pprint import pprint
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create paths to file
path_2008 = ('Resources/2008data.csv')
path_2011 = ('Resources/2011data.csv')
path_2014 = ('Resources/2014data.csv')
path_2017 = ('Resources/2017data.csv')

# Reading csv's files
medicare_data_2008_extended = pd.read_csv(path_2008)
medicare_data_2011_extended = pd.read_csv(path_2011)
medicare_data_2014_extended = pd.read_csv(path_2014)
medicare_data_2017_extended = pd.read_csv(path_2017)

medicare_data_2017_extended.head()
```

```python
# Created usuable data df by years
medicare_data_2008 = medicare_data_2008_extended.iloc[:, 0:22]
medicare_data_2011 = medicare_data_2011_extended.iloc[:, 0:22]
medicare_data_2014 = medicare_data_2014_extended.iloc[:, 0:22]
medicare_data_2017 = medicare_data_2017_extended.iloc[:, 0:22]

medicare_data_2017.head()
```

```python
# State total df broken down by years
medicare_state_data_2008 = medicare_data_2008.loc[(
                medicare_data_2008['County'] == 'STATE TOTAL'), :]

medicare_state_data_2011 = medicare_data_2011.loc[(
                medicare_data_2011['County'] == 'STATE TOTAL'), :]

medicare_state_data_2014 = medicare_data_2014.loc[(
                medicare_data_2014['County'] == 'STATE TOTAL'), :]

medicare_state_data_2017 = medicare_data_2017.loc[(
                medicare_data_2017['County'] == 'STATE TOTAL'), :]

medicare_state_data_2017
```

```python
# Converting data types from object into float (2008)
medicare_ethnicity_2008 = medicare_state_data_2008.iloc[:, [0, 8, 9, 10, 11, 12, 13]]
medicare_ethnicity_2008["Percent Female"] = medicare_ethnicity_2008["Percent Female"
                                            ].str.replace('%', '').astype(float)
medicare_ethnicity_2008["Percent Male"] = medicare_ethnicity_2008["Percent Male"
                                            ].str.replace('%', '').astype(float)
medicare_ethnicity_2008["Percent Non-Hispanic White"] = medicare_ethnicity_2008[
                "Percent Non-Hispanic White"].str.replace('%', '').astype(float)
medicare_ethnicity_2008["Percent African American"] = medicare_ethnicity_2008[
                "Percent African American"].str.replace('%', '').astype(float)
medicare_ethnicity_2008["Percent Hispanic"] = medicare_ethnicity_2008[
                "Percent Hispanic"].str.replace('%', '').astype(float)
medicare_ethnicity_2008["Percent Other/Unknown"] = medicare_ethnicity_2008[
                "Percent Other/Unknown"].str.replace('%', '').astype(float)

# Converting data types from object into float (2011)
medicare_ethnicity_2011 = medicare_state_data_2011.iloc[:, [0, 8, 9, 10, 11, 12, 13]]
medicare_ethnicity_2011["Percent Female"] = medicare_ethnicity_2011["Percent Female"
                                            ].str.replace('%', '').astype(float)
medicare_ethnicity_2011["Percent Male"] = medicare_ethnicity_2011["Percent Male"
                                            ].str.replace('%', '').astype(float)
medicare_ethnicity_2011["Percent Non-Hispanic White"] = medicare_ethnicity_2011[
                "Percent Non-Hispanic White"].str.replace('%', '').astype(float)
medicare_ethnicity_2011["Percent African American"] = medicare_ethnicity_2011[
                "Percent African American"].str.replace('%', '').astype(float)
medicare_ethnicity_2011["Percent Hispanic"] = medicare_ethnicity_2011[
                "Percent Hispanic"].str.replace('%', '').astype(float)
medicare_ethnicity_2011["Percent Other/Unknown"] = medicare_ethnicity_2011[
                "Percent Other/Unknown"].str.replace('%', '').astype(float)

# Converting data types from object into float (2014)
medicare_ethnicity_2014 = medicare_state_data_2014.iloc[:, [0, 8, 9, 10, 11, 12, 13]]
medicare_ethnicity_2014["Percent Female"] = medicare_ethnicity_2014["Percent Female"
                                            ].str.replace('%', '').astype(float)
medicare_ethnicity_2014["Percent Male"] = medicare_ethnicity_2014["Percent Male"
                                            ].str.replace('%', '').astype(float)
medicare_ethnicity_2014["Percent Non-Hispanic White"] = medicare_ethnicity_2014[
                "Percent Non-Hispanic White"].str.replace('%', '').astype(float)
medicare_ethnicity_2014["Percent African American"] = medicare_ethnicity_2014[
                "Percent African American"].str.replace('%', '').astype(float)
medicare_ethnicity_2014["Percent Hispanic"] = medicare_ethnicity_2014[
                "Percent Hispanic"].str.replace('%', '').astype(float)
medicare_ethnicity_2014["Percent Other/Unknown"] = medicare_ethnicity_2014[
                "Percent Other/Unknown"].str.replace('%', '').astype(float)

# Converting data types from object into float (2017)
medicare_ethnicity_2017 = medicare_state_data_2017.iloc[:, [0, 8, 9, 10, 11, 12, 13]]
medicare_ethnicity_2017["Percent Female"] = medicare_ethnicity_2017["Percent Female"
                                                ].str.replace('%', '').astype(float)
medicare_ethnicity_2017["Percent Male"] = medicare_ethnicity_2017["Percent Male"
                                                ].str.replace('%', '').astype(float)
medicare_ethnicity_2017["Percent Non-Hispanic White"] = medicare_ethnicity_2017[
                "Percent Non-Hispanic White"].str.replace('%', '').astype(float)
medicare_ethnicity_2017["Percent African American"] = medicare_ethnicity_2017[
                "Percent African American"].str.replace('%', '').astype(float)
medicare_ethnicity_2017["Percent Hispanic"] = medicare_ethnicity_2017[
                "Percent Hispanic"].str.replace('%', '').astype(float)
medicare_ethnicity_2017["Percent Other/Unknown"] = medicare_ethnicity_2017[
                "Percent Other/Unknown"].str.replace('%', '').astype(float)
```

```python
# Merge 2008 and 2011 ethnicity breakdown by year
ethnicity_breakdown_early = pd.merge(medicare_ethnicity_2008, medicare_ethnicity_2011, 
                        on="State", suffixes=(" 2008", " 2011"))

# Merge 2014 and 2017 tethnicity breakdown by year
ethnicity_breakdown_late = pd.merge(medicare_ethnicity_2014, medicare_ethnicity_2017, 
                        on="State", suffixes=(" 2014", " 2017"))

# Merge all ethnicity breakdowns per state
medicare_ethnicity_breakdown = pd.merge(ethnicity_breakdown_early, 
                            ethnicity_breakdown_late, on="State")

# Reformatted df
pd.options.display.float_format = '{:.2f}%'.format

# Save to csv
medicare_ethnicity_breakdown.to_csv('Cleaned_Data/medicare_ethnicity_breakdown.csv', 
                                        index = True, header = True)

medicare_ethnicity_breakdown.head()
```

```python
# Creating Total Costs per state dfs (formating them as well)

# 2008 total cost per state df
medicare_total_cost_2008 = medicare_state_data_2008.iloc[:, [0, 4, 16]]
medicare_total_cost_2008['Total Actual Costs'] = medicare_total_cost_2008[
    'Total Actual Costs'].str.replace(',', '').str.replace('$', '').astype(float)
medicare_total_cost_2008['FFS Beneficiaries'] = (
                medicare_total_cost_2008['FFS Beneficiaries'
                ].str.replace(',','').astype(float))
    
# 2011 total cost per state df
medicare_total_cost_2011 = medicare_state_data_2011.iloc[:, [0, 4, 16]]
medicare_total_cost_2011['Total Actual Costs'] = medicare_total_cost_2011[
    'Total Actual Costs'].str.replace(',', '').str.replace('$', '').astype(float)
medicare_total_cost_2011['FFS Beneficiaries'] = (
                medicare_total_cost_2011['FFS Beneficiaries'
                ].str.replace(',','').astype(float))
    
# 2014 total cost per state df
medicare_total_cost_2014 = medicare_state_data_2014.iloc[:, [0, 4, 16]]
medicare_total_cost_2014['Total Actual Costs'] = medicare_total_cost_2014[
    'Total Actual Costs'].str.replace(',', '').str.replace('$', '').astype(float)
medicare_total_cost_2014['FFS Beneficiaries'] = (
                medicare_total_cost_2014['FFS Beneficiaries'
                ].str.replace(',','').astype(float))
    
# 2017 total cost per state df
medicare_total_cost_2017 = medicare_state_data_2017.iloc[:, [0, 4, 16]]
medicare_total_cost_2017['Total Actual Costs'] = medicare_total_cost_2017[
    'Total Actual Costs'].str.replace(',','').str.replace('$', '').astype(float)
medicare_total_cost_2017['FFS Beneficiaries'] = (
                medicare_total_cost_2017['FFS Beneficiaries'
                ].str.replace(',','').astype(float))
```

```python
# Merge 2008 and 2011 total costs per state
medicare_total_cost_early = pd.merge(medicare_total_cost_2008, medicare_total_cost_2011, 
                        on="State", suffixes=(" 2008", " 2011"))

# Merge 2014 and 2017 total costs per state
medicare_total_cost_late = pd.merge(medicare_total_cost_2014, medicare_total_cost_2017, 
                        on="State", suffixes=(" 2014", " 2017"))

# Merge all total costs per state
medicare_state_costs = pd.merge(medicare_total_cost_early, 
                            medicare_total_cost_late, on="State")

# Sorted values
medicare_state_costs = medicare_state_costs.sort_values(
                            by = 'Total Actual Costs 2008', ascending = False)

# Reformatted df
pd.options.display.float_format = '{:,.0f}'.format
medicare_state_costs = medicare_state_costs.rename(
                            columns = {'FFS Beneficiaries 2008':'Beneficiaries 2008',
                            'FFS Beneficiaries 2011':'Beneficiaries 2011',
                            'FFS Beneficiaries 2014':'Beneficiaries 2014',
                            'FFS Beneficiaries 2017':'Beneficiaries 2017'})

# Save to csv
medicare_state_costs.to_csv('Cleaned_Data/medicare_state_costs.csv', 
                                        index = True, header = True)


medicare_state_costs.style.format({'Total Actual Costs 2008':'${:,.2f}'.format,
                        'Total Actual Costs 2011':'${:,.2f}'.format,
                        'Total Actual Costs 2014':'${:,.2f}'.format,
                        'Total Actual Costs 2017':'${:,.2f}'.format, 
                        'Beneficiaries 2008': '{:,.0f}'.format,
                        'Beneficiaries 2011': '{:,.0f}'.format,
                        'Beneficiaries 2014': '{:,.0f}'.format,
                        'Beneficiaries 2017': '{:,.0f}'.format,})
```
